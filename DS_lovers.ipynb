# Donut chart()
# 1.labels are for the differencation of the chart
# 2.colors are used for coloring and showing variance between frequency
# 3.strangle is for angle for slicing
# 4.shadow is true means it will drop the chart
# 5.radius is to set radius of circle
# 6.explode is given for distance of % from the centre or to cut the parts from the pie
# 7.autopct is to show the percentage of the chart 
# 8.pctdistance moves the cut piece somewhat far from the chart

# using plt.circle we can create a circle

```python
#plt.gcf() is uesd to current figure


plt.pie(a,labels=['Not churned',"churned"],colors=['red','blue'] , startangle=50 , shadow=True , radius =2 ,
       explode=(0,0.2) ,autopct='%1.2f%%' , pctdistance=0.75)



plt.title("churn vs non churn customers")
```

```python
circle=plt.Circle((0,0),radius=1,color='green')

c=plt.gcf()

c.gca().add_artist(circle)
```

Count plot

```python
plt.figure(figsize=(12,10))  #Used for increasing the picture size
sns.countplot(data['state'])

plt.xticks(fontsize=10)      #Use this funstion for increasing the size of the x lable
plt.yticks(fontsize=10)      #Use this funstion for increasing the size of the y lable

plt.ylabel('States',fontsize=20,color='red')   #Accessing label x
plt.xlabel('Counts',fontsize=20,color='Green') #Accessing label y
plt.show()
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(12,10))
sns.countplot(x=data.state, hue=data.churn)   # we are using Hue for color coding for the column churn
                                              # this says that the color diifers for different values in it 
                                              
plt.xlabel('State',fontsize=10,color='blue')
plt.ylabel('Counts',fontsize=10,color='orange')

plt.show()
```
