Bins - gives the leverage of the changing the range in the graph 


histogram bins along the range of the data and then drawing bars to show the number of conservation that fall in each bin

```python
df.describe()
```

```python
bins=[2,3,4,5,6,7,8,9,10]
sns.distplot(df['time_spend_company'])
plt.xticks(bins)
```

```python
sns.distplot(df['time_spend_company'], rug=True)
plt.xticks(bins)
```

```python
sns.distplot(df['time_spend_company'], bins=bins , rug=True , hist_kws={'color': 'blue', 'edgecolor':'red','linewidth':3,'alpha':1})
plt.xticks(bins)
```

```python
sns.distplot(df['time_spend_company'], bins=bins ,
             rug=True ,
             hist_kws={'color': 'blue', 'edgecolor':'red','linewidth':3,'alpha':1})
plt.xticks(bins)
```
