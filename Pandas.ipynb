# Merge

```python
df1=pd.DataFrame({'Roll_no':[1,2,3,4,5],'Physics':[6,7,8,9,10]})
df1
```

```python
df2=pd.DataFrame({'Roll_no':[1,2,3,4,5],'Chemistry':[11,12,13,14,15]})
df2
```

```python
pd.merge(df1,df2,on='Roll_no')
```

```python
pd.merge(df2,df1,on = 'Roll_no')
```

```python
df3=pd.DataFrame({'Roll_no:':[12,3,4,5,6],'Physics':[12,87,67,4,3]})
df4=pd.DataFrame({'Roll_no:':[12,3,8,9,6],'Chemistry':[12,87,67,4,3]})

pd.merge(df3,df4)
```

```python
pd.merge(df3,df4, how = 'left')
```

```python
pd.merge(df3,df4,how = 'right')
```

```python
pd.merge(df3,df4, how = 'outer')
```
